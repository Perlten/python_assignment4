# Part 1
Create a line graph showing the development in poverty over time in both Argentina, Ivory cost, USA. (be aware of the missing data in particular years)

```python
# Filters all irrelevant countries
c_data = data.loc[(data["Country Name"].isin(
    ["Argentina", "United States", "Venezuela, RB", "Denmark"]))]
# Filter all irrelevant data
c_data = c_data.iloc[:, 4:]
# Creates labels for legend
label = [contries_names[x] for x in c_data.index]
module.createDataFramePlot(c_data.values[:,:-1], c_data.columns[:-1], label, style="o-")
```

# Part 2
Find the 10 countries with the highest poverty rate, and plot the rates in a bar chart

```python
year = "2000"

# Get the ten countries with the largest poverty rate in a given year 
c_data = data.sort_values(by=[year], ascending=False).head(10)
# Filters away all data in irrelevant years
c_data = c_data.loc[:, year]
# Create x-axis 
index = [contries_names[x] for x in c_data.index]
module.createDataFramePlot([c_data.values], index, x_label="Year", plt_type="bar",
                    y_label="Poverty Rate", title="Poverty Rate in {}".format(year))
```
